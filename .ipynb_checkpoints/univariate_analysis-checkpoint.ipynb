The relationship here is "clearly decreasing with days to maturity." The relation also appears to be monotonic, but also not linear in relationship.

```python
fig, ax = plt.subplots(figsize = (16,9))
df.groupby('daysToExpiry')['volume'].mean().plot()
```

The open interest, on the other hand is clearly linearly decreasing, but only within bands. At certain points it spikes.
```python
df_spx = pd.read_csv('./dat/spx_index_prices.csv')
```

```python
df_spx.head()
```

```python
df_spx['date'] = pd.to_datetime(df_spx['Date'])
del df_spx['Date']
```

```python
df = df.merge(df_spx, on = 'date', how = 'left', validate = 'm:1')
```

```python
df['moneyness'] = df['strike_price']/df['Close'] / 10 - 100
df['moneyness'].describe().iloc[1:]
```

At the money options are the most volatile simply because they are the most sensitive to moves in the underlying price (which is the primary stochastic factor driving options prices). Thus, to get risk exposure to this asset class it makes sense that more traders would want to trade at the money options than out of or in the money options. The distribution is a bit wider for the out of the money options (strike less than spot). That is intuitive, because more people trade out of the money options as a hedge for larger and larger declines. The important thing to note however, is that there is an imbalance in the dependence for negative vs. positive moneyness, suggesting simple linear relationships simply will not cut it. 

```python
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df['moneyness'], df['volume'], alpha = 0.05)
```

On the other hand for sorts on moneyness, the open interest has a wider distribution. That makes sense because although most people would trade at the money options, there's no reason for the amount of outstanding options to concentrate any more heavily on 0. 

```python
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df['moneyness'], df['open_interest'], alpha = 0.05)
```

It does not directly seem that the relationship between option volume and underlying volume is positive. One could argue it staring at the plot long enough.

```python
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df['Volume'], df['volume'], alpha = 0.05)
```

But one problem with this one is that it starts to chunk. That was actually expected because the data is a panel dataset so the value on the x-axis will be the same for a particular date. The variation in the y-axis just reflects the variation in volume across the set of strikes and life times of the options in the dataset. 

Averaging across the chain for each date allows us to get another sense of the non-linearities here. In particular, when the underlying volume is less than about 2 billion shares, the relationship between option volume and underlying volume has a very steep slope. When it exceeds this threshold, that slope flattens immensely. 

```python
df_plt = df.groupby('date')[['volume','Volume']].mean()
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df_plt['Volume'] / 1e9, df_plt['volume'], alpha = 0.3)
```

A similar effect happens when we look at open interest vs. underlying volume. There's a wider distribution in the scatter plot and there are clearly two regims relating the variables when averaged across the option chain. 

```python
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df['Volume'], df['open_interest'], alpha = 0.05)
```

```python
df_plt = df.groupby('date')[['open_interest','Volume']].mean()
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df_plt['Volume'] / 1e9, df_plt['open_interest'], alpha = 0.3)
```

# Regression

```python
def get_mean_series(g):
    g = g.sort_values('date')
    vals = g['volume'].expanding().mean().values
    dats = g['date'].values
    return pd.DataFrame({'date':dats, 'roll_avg_vlm':vals})
```

```python
slct = ['exdate','strike_price','lifeTime']
df_adv = df.groupby(slct).apply(get_mean_series).reset_index(level = slct)
```

```python
df_adv.head()
```

```python
df_adv['date']+= pd.Timedelta('1 days')
```

```python
df_adv = df_adv.sort_values(['date']+slct).reset_index(drop = True)
```

```python
df2 = pd.merge_asof(df, df_adv, on = "date", by = slct)
```

```python
len(df2)
```

```python
df2.head()
```

```python
idx = df2['volume'] > 0
```

```python
fig, ax = plt.subplots(figsize = (16,9))
plt.scatter(df2[idx]['roll_avg_vlm'], df2[idx]['volume'], alpha = 0.3)
```

```python
import statsmodels.api as sm
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm', data = df2).fit().summary()
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm + Volume', data = df2).fit().summary()
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm + I(abs(moneyness))', data = df2).fit().summary()
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm + I(Volume/1e9) + I(abs(moneyness))', data = df2).fit().summary()
```

```python
df2['BigVolume'] = (df['Volume'] > 3e9).astype(int)
df2['x1'] = df['Volume'] * (1 - df2['BigVolume']) / 1e9
df2['x2'] = df['Volume'] * df2['BigVolume'] / 1e9
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm + BigVolume + x1 + x2 + I(abs(moneyness))', data = df2).fit().summary()
```

```python
sm.OLS.from_formula('volume ~ roll_avg_vlm + BigVolume + x1 + x2 + I(abs(moneyness))', data = df2).fit().summary()
```
